**b. Conversion from Applications & ISIR Applications**

```python
import pandas as pd

sdata_url = 'https://raw.githubusercontent.com/arman-hossain-git/admissions/main/data/Students.csv'
dfs = pd.read_csv(sdata_url)

dfs.head()
```

```python
dfs.columns
```

```python
# Convert date columns to datetime format
# Fill missing values  for required columns

dates = ['Timeline', 'Demographic_Timeline', 'startDate', 'endDate', 'midPointDate', 'graduationDate', 'LDA', 'cohortStartDate', 'dateOfBirth', 'Conversion_Date', 'creationDtTm']
for col in dates:
    dfs[col] = pd.to_datetime(dfs[col], errors='coerce')


req = ['city', 'state', 'studentCampus', 'sex', 'race', 'expectedFamilyContribution', 'isFirstTimeStudent', 'livingArrangements', 'incomeLevel', 'residencyStatus', 'familySize', 'TEAM_NAME', 'remarks']
for col in req:
  dfs[col] = dfs[col].fillna('UNSPECIFIED')
```

```python
dfs.head()
```

```python
# Create a new column for campusCode

# Reference
# campusCode = 34652 --> Campus = 1. NEW PORT RICHEY
# campusCode = 34601 --> Campus = 2. BROOKSVILLE
# campusCode = 34606 --> Campus = 3. SPRING HILL

def campus(code):
  if code == 34652:
    return '1. NEW PORT RICHEY'
  elif code == 34601:
    return '2. BROOKSVILLE'
  elif code == 34606:
    return '3. SPRING HILL'
  elif code == -1:
    return 'UNSPECIFIED'
  elif code == 0:
    return 'UNSPECIFIED'
  else:
    'UNSPECIFIED'

dfs['campus'] = dfs['studentCampus'].apply(campus)
dfs[['studentCampus', 'campus']].head()
```

```python
dfs[['studentCampus']].value_counts()
```

```python
dfs['campus']
```

```python
dfs['sex'].value_counts()
```

```python
# Create a new column for gender

# For all empty/null/NaN value in column sex we want 'UNSPECIFIED'

def null_gender(code):
  if code == 1:
    return '1. MALE'
  elif code == 2:
    return '2. FEMALE'
  elif code == 0:
    return 'OTHER'
  elif code == 4:
    return 'OTHER'
  else:
    'OTHER'

dfs['gender'] = dfs['sex'].apply(null_gender)
dfs[['sex', 'gender']].head()
```

```python
# Create a new column for ethnicity

# For all empty/null/NaN value in column race we want 'UNSPECIFIED'

def null_ethnicity(code):
  if code == 1:
    return '1. White'
  elif code == 2:
    return '2. African American'
  elif code == 3:
    return '3. Two or more races'
  elif code == 4:
    return '4. Unknown Other'
  elif code == 5:
    return '5. Asian'
  elif code == 6:
    return '6. Hispanic'
  elif code == 7:
    return '7. Alaska Native'
  elif code == 8:
    return '8. American Indian'
  elif code == 0:
    return 'UNSPECIFIED'
  else:
    'UNSPECIFIED'

dfs['ethnicity'] = dfs['race'].apply(null_ethnicity)
dfs[['race', 'ethnicity']].head()
```

```python
dfs['ethnicity'].value_counts()
```

```python
dfs['isFirstTimeStudent'].value_counts()
```

```python
# Create a new column for first time students

# For all empty/null/NaN/-1 value in column isFirstTimeStudent we want 'UNSPECIFIED'

def null_isFirstTimeStudent(code):
  if code == 1:
    return '1. YES'
  elif code == 0:
    return '2. NO'
  elif code == 'UNSPECIFIED' or code == -1:
    return 'UNSPECIFIED'
  else:
    'UNSPECIFIED'

dfs['FirstTime_Student'] = dfs['isFirstTimeStudent'].apply(null_isFirstTimeStudent)
dfs[['isFirstTimeStudent', 'FirstTime_Student']].head()
```

```python
# Create a sub dataframe consisting of the students who either enrolled or started after 2020-01-01 and their enroll type is in (Enrolled", "Graduated", "Withdrawn", "LOA", "Deceased")

sub_df = dfs[(dfs['Demographic_Timeline'] >= '2020-01-01') & (dfs['Student_Enroll_Type'].isin(["Enrolled", "Graduated", "Withdrawn", "LOA", "Deceased"]))]
```

```python
# Number of students who either enrolled or started a program after '2020-01-01'

number_of_student = sub_df['studentId'].nunique()

number_of_student
```

```python
# Number of students who started a program after '2020-01-01'

number_of_strats = sub_df[(sub_df['registrationId'].notnull())]['studentId'].nunique()

number_of_strats
```

```python
# Number of students who exclusively enrolled from Application or ISIR after '2020-01-01'

number_of_appisir = sub_df[
    (sub_df['creationDtTm'] >= '2020-01-01') &
    (
        sub_df['runningCumStuStatus'].str.contains(r"1. ISIR", na=False) |
        sub_df['runningCumStuStatus'].str.contains(r"1. Appli", na=False)
    )
]['studentId'].nunique()

number_of_appisir
```

```python
# Application & ISIR rate

rate_of_appisir = number_of_appisir/number_of_student*100

rate_of_appisir
```

```python
# Starts from Application & ISIR

appisir_start = sub_df[
    (sub_df['creationDtTm'] >= '2020-01-01') &
    (
        sub_df['runningCumStuStatus'].str.contains(r"1. ISIR", na=False) |
        sub_df['runningCumStuStatus'].str.contains(r"1. Appli", na=False)
    ) &
    (sub_df['registrationId'].notnull())
]['studentId'].nunique()

appisir_start
```

```python
# Start rate from Application & ISIR

appisir_start_rate = appisir_start/number_of_appisir*100

appisir_start_rate
```
